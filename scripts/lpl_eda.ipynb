```python
from datetime import date

import altair as alt
import polars as pl

alt.data_transformers.enable("vegafusion")
```

```python
raw = pl.scan_parquet("../data/raw.parquet")
raw_state = raw.filter(pl.col("geography_type") != "nation").collect()
raw_state.write_parquet(file="../data/raw_state.parquet")
```

```python
score = pl.read_parquet("../output/flu_state_all/mspe.parquet")
score = score.sort(["season"])
```

```python
alt.Chart(score.filter(pl.col("forecast_date") == date(2022, 9, 1))).mark_rect().encode(
    alt.X("geography"), alt.Y("season"), alt.Color("score_value", title="MSPE")
)
```
